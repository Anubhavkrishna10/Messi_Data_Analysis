# 3.Data Cleaning
1. Imputing the missing values
2. Removing unimportant columns

```python
messi.drop('competition_name',axis=1,inplace=True)
```

```python
messi.drop('goal_type',axis=1,inplace=True)
```

```python
round(messi.isnull().sum()*100/len(messi),2)
```

```python
messi.home_team.value_counts()
```

```python
messi.minute.value_counts()
```

```python
def get_half(x):
    if '+' in x:
        return 'extra_time'
    elif int(x)<46:
        return 'first_half'
    else:
        return 'second_half'

messi['goal_halves']=messi.minute.apply(lambda x:get_half(x))
messi.head()

```

```python
def transform(y):
    if '+' in y:
        
        sum1= y.split('+')
        return int(sum1[0])+int(sum1[1])
        
    else:
        return int(y)
    
messi.minute=messi.minute.apply(lambda y:transform(y))
messi.minute.astype(int)
messi.head()
```

```python
def win_not_win(z):
    if 'Won' in z:
        return 1
    else:
        return 0
messi['win_not_win']=messi.final_game_result.apply(lambda z:win_not_win(z))
messi.head()
```

```python
messi.win_not_win.value_counts()
```

# 4.Data Analysis

```python
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns



plt.figure(figsize=(100,100))

sns.set(font_scale = 5)

plt.subplot(3,2,1)
ax=sns.countplot(messi.win_not_win)
for rect in ax.patches:
    ax.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 50)


plt.subplot(3,2,2)
bx=sns.countplot(messi.final_game_result)
for rect in bx.patches:
    bx.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 50)


plt.subplot(3,2,3)
cx=sns.countplot(messi.goal_method,hue=messi.win_not_win)
for rect in cx.patches:
    cx.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 50)

plt.subplot(3,2,4)
dx=sns.countplot(messi.goal_halves,hue=messi.win_not_win)
for rect in dx.patches:
    dx.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 50)

plt.subplot(3,2,5)
ex=sns.countplot(messi.score_away_team_when_messi_scored,hue=messi.win_not_win)
for rect in ex.patches:
    ex.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 50)

plt.subplot(3,2,6)
fx=sns.countplot(messi.score_home_team_when_messi_scored,hue=messi.win_not_win)
for rect in fx.patches:
    fx.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 50)




plt.show()
```

* In Messi's career in which he scored,he has **won 586 matches** ,**drawn 47 matches** and **lost in 18 matches** while playing for his club and country.

```python



plt.figure(figsize=(50,120))
sns.set(font_scale = 4)
cx=sns.countplot(y = messi.opponent,hue=messi.goal_halves)
cx.bar_label(cx.containers[1])
cx.bar_label(cx.containers[0])
plt.title("Opponent vs Goal_scored_halves")
plt.legend(loc ="center right")
plt.show()


```

# Performance Against Top Spanish Clubs
*  In Laliga against top clubs,Messi scored more goals in first half against **Sevilla**.
*  His performance against **Real Madrid** is better in second half than first half.
*  Against **Atletico Madrid** his performance is equally good in both halves.

```python
plt.figure(figsize = (15,15))
a = 1

for method in messi.goal_method.value_counts().index.values:
    plt.subplot(2,3,a)

    data = messi[messi.goal_method == method].groupby(['goal_halves'])['goal_halves'].count().reset_index(name = 'Count')
    textprops = {"fontsize":15}
    plt.pie(data.Count,autopct='%1.1f%%',textprops =textprops )
    plt.title(method,fontsize = 15)
    plt.legend(title = 'Goal',labels= data.goal_halves,fontsize =7,title_fontsize =7)
    
    a += 1
plt.show()
```

# Observations
* Goals scored using left foot are more in second half than first half.
* Goals scored using right foot are more in second half than first half.
* In all his right footed goals the percentage difference in goals of second half and first half is more than in  percentage difference of goals scored in second half and first half in all his left footed goals.
* In his career he has equally scored in both halves of game using his head.
* He has only scored in second half using his chest and only scored in first half using his hand and hip.

```python
plt.figure(figsize = (6,6))
teams = ["FC Barcelona","Argentina"]
a = 1
for team in teams:
    plt.subplot(2,2,a)

    data = messi[messi.home_team == team].groupby(['win_not_win'])['win_not_win'].count().reset_index(name = 'Count')
    textprops = {"fontsize":15}
    plt.pie(data.Count,autopct='%1.1f%%',textprops =textprops )
    plt.title(team,fontsize = 15)
    plt.legend(title = "result",labels= data.win_not_win,fontsize = 10,title_fontsize =10)
    a += 1    
    
plt.show()
```

* Winning percentage of FC Barcelona at home is less than winning percentage of Argentina at home.
* Overall the winning percentage for both Barcelona and Argentina is quite high with **92.4%** and **94.7% **respectively.

```python
plt.figure(figsize = (6,6))
teams = ["FC Barcelona","Argentina"]
a = 1
for team in teams:
    plt.subplot(2,2,a)

    data = messi[messi.away_team == team].groupby(['win_not_win'])['win_not_win'].count().reset_index(name = 'Count')
    textprops = {"fontsize":15}
    plt.pie(data.Count,autopct='%1.1f%%',textprops =textprops )
    plt.title(team,fontsize = 15)
    plt.legend(title = "result",labels= data.win_not_win,fontsize = 10,title_fontsize = 10)
    a += 1    
    
plt.show()
```

* Winning percentage of FC Barcelona in away games is less than winning percentage of Argentina at away games.
* Overall the winning percentage for both Barcelona and Argentina in away games is decent  with **86%** and **88.9%** respectively but less than when compared to their performance at home.
